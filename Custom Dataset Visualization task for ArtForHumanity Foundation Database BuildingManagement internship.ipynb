```python
import numpy as np
import matplotlib.pyplot as plt

# Data
states = ['NY', 'CA', 'IL', 'FL']
organization_types = ['Finance', 'Education', 'NGO', 'Technology', 'Healthcare', 'Retail', 'Hospitality', 'Non-Profit', 'Manufacturing', 'Consulting']
organization_counts = [
    [1, 0, 0, 0],  # NY
    [1, 1, 0, 0],  # CA
    [0, 0, 1, 0],  # IL
    [0, 0, 0, 1]   # FL
]

# Transpose organization counts
organization_counts = np.array(organization_counts).T.tolist()

# Plotting
plt.figure(figsize=(10, 6))

bar_width = 0.2
index = np.arange(len(states))

for i in range(len(states)):  # Iterate over the states
    plt.bar(index + i * bar_width, organization_counts[i], bar_width, label=organization_types[i])

plt.xlabel('State')
plt.ylabel('Number of Organizations')
plt.title('Number of Organizations by State and Type')
plt.xticks(index + bar_width*4.5, states)
plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
```
